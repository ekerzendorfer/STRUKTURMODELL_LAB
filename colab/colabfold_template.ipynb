# ColabFold-Workflow für das KI-Strukturmodell-Labor v0.4.0

Dieses Notebook ist eine **externe Arbeitshilfe**. Die Webseite faltet nicht live, sondern vergleicht vorbereitete PDB-Dateien.

> Sequenz vorbereiten → in ColabFold falten → mehrere PDB-Modelle herunterladen → im Webtool testen → bestes und alternatives Modell ins Repo übernehmen.


## Ziel-Dateinamen ab v0.4.0

Für jedes Beispiel können zwei KI-Modelle hinterlegt werden:

```text
af2_best.pdb         # bestbewertetes ColabFold-Modell
af2_alternative.pdb  # niedriger bewertetes Vergleichsmodell
```

Das Vergleichsmodell wird bewusst neutral benannt: Es ist nicht automatisch „falsch“, sondern dient zur Diskussion von Bewertung, Unsicherheit und Modellgrenzen.


````python
examples = {
    'trp_cage': {
        'title': 'Trp-cage',
        'sequence': 'NLYIQWLKDGGPSSGRPPPS',
        'best_path': 'structures/trp_cage/af2_best.pdb',
        'alternative_path': 'structures/trp_cage/af2_alternative.pdb',
        'didactic_note': 'Mini-Protein: kurze Sequenz, bewegliches System, guter Einstieg.'
    },
    'ubiquitin': {
        'title': 'Ubiquitin',
        'sequence': 'MQIFVKTLTGKTITLEVEPSDTIENVKAKIQDKEGIPPDQQRLIFAGKQLEDGRTLSDYNIQKESTLHLVLRLRGG',
        'best_path': 'structures/ubiquitin/af2_best.pdb',
        'alternative_path': 'structures/ubiquitin/af2_alternative.pdb',
        'didactic_note': 'Referenzprotein: stabiler Fold, sehr guter Vergleich KI-Modell vs. Experiment.'
    },
    'calmodulin': {
        'title': 'Calmodulin',
        'sequence': 'ADQLTEEQIAEFKEAFSLFDKDGDGTITTKELGTVMRSLGQNPTEAELQDMINEVDADGNGTIDFPEFLTMMARKMKDTDSEEEIREAFRVFDKDGNGYISAAELRHVMTNLGEKLTDEEVDEMIREADIDGDGQVNYEEFVQMMTAK',
        'best_path': 'structures/calmodulin/af2_best.pdb',
        'alternative_path': 'structures/calmodulin/af2_alternative.pdb',
        'didactic_note': 'Calcium-bindendes Protein: Sequenzmodell vs. konkreter Ca²⁺-gebundener experimenteller Zustand.'
    }
}

print('Verfügbare Beispiele:')
for key, item in examples.items():
    print(f'- {key}: {item["title"]} → {item["didactic_note"]}')
````

## 1. Beispiel auswählen

Ändere bei Bedarf nur die folgende Variable.


````python
selected = 'calmodulin'  # 'trp_cage', 'ubiquitin' oder 'calmodulin'
item = examples[selected]
print('Ausgewählt:', item['title'])
print('Sequenzlänge:', len(item['sequence']), 'Aminosäuren')
print('Didaktischer Hinweis:', item['didactic_note'])
print('Zielpfad bestes Modell:', item['best_path'])
print('Zielpfad Vergleichsmodell:', item['alternative_path'])
````

## 2. FASTA-Block kopieren

Kopiere den folgenden FASTA-Block in das externe ColabFold/AlphaFold2-Notebook. Die zusätzlich geschriebene `.fasta`-Datei ist nur eine lokale Hilfe in diesem Notebook.


````python
fasta_text = f">{item['title']}\n{item['sequence']}\n"
print('--- FASTA-BLOCK ZUM KOPIEREN ---')
print(fasta_text)
with open(f"{selected}.fasta", 'w', encoding='utf-8') as f:
    f.write(fasta_text)
print('FASTA-Datei zusätzlich geschrieben:', f'{selected}.fasta')
````

## 3. Externes ColabFold/AlphaFold2-Notebook öffnen

[ColabFold AlphaFold2 Notebook öffnen](https://colab.research.google.com/github/sokrypton/ColabFold/blob/main/AlphaFold2.ipynb)

Dort den FASTA-Block einfügen und mit Standardparametern rechnen lassen. Für die beiden einfachen Beispiele genügt zunächst Einzelkette/Monomer.


## 4. PDB-Dateien auswählen und benennen

Nach dem Lauf erzeugt ColabFold mehrere gerankte Modelle.

- höchstbewertetes Modell → `af2_best.pdb`
- niedriger bewertetes Modell → `af2_alternative.pdb`

Empfohlen: zuerst beide PDBs im Webtool über **Eigenes PDB testen** ansehen. Danach ins Repo legen als:

```text
{item['best_path']}
{item['alternative_path']}
```

Danach GitHub Pages kurz deployen lassen und die Webseite mit Strg+F5 neu laden.


## Optional: didaktisches Störmodell ab v0.4.1

Zusätzlich kann pro Beispiel eine Datei `didactic_decoy.pdb` abgelegt werden:

```text
structures/trp_cage/didactic_decoy.pdb
structures/ubiquitin/didactic_decoy.pdb
```

Diese Datei ist **kein AlphaFold/ColabFold-Ergebnis**, sondern ein bewusst gewähltes Vergleichsmodell, z. B. aus PEP-FOLD, QUARK, Phyre2, I-TASSER oder einer transparent künstlich veränderten Struktur.

Fehlt diese Datei, zeigt die Webseite nur einen Hinweis und lädt weiter die vorhandenen Standardmodelle.


## Hinweis zu Calmodulin ab v0.5.0

Für Calmodulin wird bewusst die 148-AS-Sequenz passend zur experimentellen Struktur 1CLL verwendet. Diese Sequenz beginnt mit `ADQL...`, also ohne zusätzliches initiierendes Methionin. Dadurch bleiben die Residuenummern im einfachen Viewer-Vergleich besser vergleichbar.

Die ColabFold-Modelle enthalten keine expliziten Calcium-Ionen. Genau dieser Unterschied ist didaktisch erwünscht: Das experimentelle Modell zeigt einen konkreten Ca²⁺-gebundenen Zustand.
