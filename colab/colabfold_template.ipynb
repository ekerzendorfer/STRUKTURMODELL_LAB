# ColabFold-Workflow für das KI-Strukturmodell-Labor

Dieses Notebook ist in v0.1 noch bewusst einfach gehalten. Es soll den KI-Prozess sichtbar machen:

> Sequenz → ColabFold / AlphaFold2 → PDB-Datei → Vergleich im KI-Strukturmodell-Labor

Für den Unterricht bleibt die Website die stabile Oberfläche. Dieses Notebook ist die optionale KI-Vertiefung.


## 1. Sequenz auswählen

Die folgenden Sequenzen sind für die ersten Beispiele vorbereitet.


```python
examples = {
    'Trp-cage': 'NLYIQWLKDGGPSSGRPPPS',
    'Ubiquitin': 'MQIFVKTLTGKTITLEVEPSDTIENVKAKIQDKEGIPPDQQRLIFAGKQLEDGRTLSDYNIQKESTLHLVLRLRGG'
}

for name, seq in examples.items():
    print(f'>{name}')
    print(seq)
    print()
```

## 2. FASTA-Datei schreiben

Wähle ein Beispiel und schreibe die Sequenz als FASTA-Datei. Diese Datei kann in ColabFold verwendet werden.


```python
selected = 'Ubiquitin'  # 'Trp-cage' oder 'Ubiquitin'
seq = examples[selected]
fasta_path = f'{selected.replace(" ", "_").lower()}.fasta'
with open(fasta_path, 'w', encoding='utf-8') as f:
    f.write(f'>{selected}\n{seq}\n')
print(f'FASTA geschrieben: {fasta_path}')
```

## 3. ColabFold starten

Für eine einfache Live-Faltung kann das öffentliche ColabFold-Notebook verwendet werden:

[ColabFold AlphaFold2 Notebook öffnen](https://colab.research.google.com/github/sokrypton/ColabFold/blob/main/AlphaFold2.ipynb)

Dort die Sequenz einfügen und nach der Berechnung die beste PDB-Datei herunterladen.

Langfristig kann diese Datei im Webtool über **Eigenes PDB testen** geladen oder als vorbereitete Struktur in das Repo übernommen werden.
