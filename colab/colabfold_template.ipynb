# ColabFold-Workflow für das KI-Strukturmodell-Labor

Dieses Notebook ist die **externe KI-Vertiefung** zum Webtool.

Didaktischer Ablauf:

1. Sequenz auswählen
2. ColabFold/AlphaFold2-Modell erzeugen
3. beste oder bewusst auch schlechter bewertete PDB-Datei herunterladen
4. PDB im Webtool über **Eigenes PDB testen** importieren
5. später kuratiert unter `structures/<beispiel>/af2_colabfold.pdb` ins Repo übernehmen

Die Website bleibt bewusst stabil und führt keine Live-Faltung aus.


## 1. Sequenz auswählen

Die ersten Beispiele sind vorbereitet. Die Dateinamen sind so gewählt, dass sie direkt zur Repo-Struktur passen.


```python
examples = {
    'trp_cage': {
        'title': 'Trp-cage',
        'sequence': 'NLYIQWLKDGGPSSGRPPPS',
        'target_path': 'structures/trp_cage/af2_colabfold.pdb'
    },
    'ubiquitin': {
        'title': 'Ubiquitin',
        'sequence': 'MQIFVKTLTGKTITLEVEPSDTIENVKAKIQDKEGIPPDQQRLIFAGKQLEDGRTLSDYNIQKESTLHLVLRLRGG',
        'target_path': 'structures/ubiquitin/af2_colabfold.pdb'
    }
}

for key, item in examples.items():
    print(key, '→', item['title'])
    print(item['sequence'])
    print('Zielpfad im Repo:', item['target_path'])
    print()
```

## 2. FASTA-Datei schreiben

Wähle eines der Beispiele. Die erzeugte FASTA-Datei kann in ColabFold verwendet werden.


```python
selected = 'ubiquitin'  # 'trp_cage' oder 'ubiquitin'
item = examples[selected]
fasta_path = f'{selected}.fasta'

with open(fasta_path, 'w', encoding='utf-8') as f:
    f.write(f">{item['title']}\n{item['sequence']}\n")

print('FASTA geschrieben:', fasta_path)
print('Nach der Faltung beste PDB-Datei im Repo ablegen als:')
print(item['target_path'])
```

## 3. ColabFold starten

Für die eigentliche Faltung wird das öffentliche ColabFold-Notebook verwendet:

[ColabFold AlphaFold2 Notebook öffnen](https://colab.research.google.com/github/sokrypton/ColabFold/blob/main/AlphaFold2.ipynb)

Dort die Sequenz einfügen. Nach der Berechnung können die PDB-Dateien aus den Ergebnissen heruntergeladen werden.

Für das Webtool sind zwei Varianten didaktisch interessant:

- **bestes Modell**: Standardvergleich mit Experiment
- **schlechter bewertetes Modell**: Vertiefung, um Konfidenz, Modellgrenzen und Bewertungslogik zu diskutieren


## 4. Ergebnis in das Webtool bringen

### Variante A: Sofort testen

PDB-Datei herunterladen und im Webtool über **Eigenes PDB testen** importieren.

### Variante B: Standardmodell ins Repo übernehmen

Die beste, kuratierte PDB-Datei passend umbenennen:

- Trp-cage: `structures/trp_cage/af2_colabfold.pdb`
- Ubiquitin: `structures/ubiquitin/af2_colabfold.pdb`

Danach im Repo hochladen. Die Website lädt diese Datei dann automatisch im Standardmodus.
